Carilah derivative/turunan dari sebuah fungsi:
 * $y = x^{2} + 2x + 1 \\ $

## Symbolic

```python
#cara symbolic
import sympy as sy

x = sy.Symbol('x',real=True)
```

```python
y1 = x**2+2*x+1

y1.diff(x)
```

```python
y2 = 4*x**3+3*x**2+2*x+1

y2.diff(x)
```
