### Measure of Central Tendency

```python
#Get the Mean and Median

print("{} \t{} \t{}".format("Column Name","Mean", "Median").expandtabs(30))
print("-"*80)
for col in covid_19.select_dtypes("number"):
    print("{} \t{} \t{}".format(col, covid_19[col].mean(), covid_19[col].median()).expandtabs(30))
```

```python
#Get the mode
print("{} \t{}".format("Column Name", "Mode").expandtabs(30))
print("-"*50)
for col in covid_19.select_dtypes('number'):
    print("{} \t{}".format(col, covid_19[col].mode()).expandtabs(30))
```

```python
#for categorical column
cov = covid_19.select_dtypes('object')

cov['Special Status'].mode()
  
```

```python
# for categorical column
print("{} \t{}".format("Column Name", "Mode").expandtabs(30))
print("-"*50)
for col in covid_19.select_dtypes("object"):
    print("{} \t{}".format(col, covid_19[col].mode()).expandtabs(30))
```

### Measure of Variance

```python
covid_19.head()
```

#### Range

```python
cov_num = covid_19.select_dtypes('number')
```

```python
cov_num.head()
```

```python
print("{} \t{}".format("Column Name", "Range").expandtabs(30))
print("-"*50)
for col in cov_num:
    print("{} \t{}".format(col, cov_num[col].max()-cov_num[col].min()).expandtabs(30))
```

#### Percentile and Quartiles

##### Percentiles

```python
from scipy import stats

print('{} \t{}'.format('Column Names', 'Data with Index 40 and strict percentile ').expandtabs(30))
print("-"*60)
for col in cov_num:
    print('{} \t{}'.format(col, stats.percentileofscore(cov_num[col], cov_num[col][40], 'strict')).expandtabs(30))
```

```python
print('{} \t{}'.format('Column Names', 'Data with Index 40 and rank percentile ').expandtabs(30))
print("*"*80)
for col in cov_num:
    print('{} \t{}'.format(col, stats.percentileofscore(cov_num[col], cov_num[col][40], 'rank')).expandtabs(30))
```

##### Quartile

```python
print('{} \t{} \t{}'.format('Column Names', 'Q1', 'Q2').expandtabs(30))
print('*'*80)
for col in cov_num:
    print("{} \t{} \t{}".format(col, cov_num[col].quantile(0.25), cov_num[col].quantile(0.5)).expandtabs(30))
```

```python
print('{} \t{}'.format('Column Names', 'Q3').expandtabs(30))
print('*'*50)
for col in cov_num:
    print("{} \t{}".format(col, cov_num[col].quantile(0.75)).expandtabs(30))
```

#### Variance & Standard Deviation

```python
print("{} \t{} \t{}".format('Column Names','Variance', 'Standard Deviation').expandtabs(30))
print('*'*80)
for col in cov_num:
    print('{} \t{} \t{}'.format(col, cov_num[col].var(), cov_num[col].std()).expandtabs(30))
```

#### Z Score

```python
print("{} \t{} \t{}".format("Column Name", "data[col][0]", "Z Score").expandtabs(30))
print("-"*80)
for col in cov_num:
    print("{} \t{} \t{}".format(col, cov_num[col][0], (cov_num[col][0]-cov_num[col].mean())/cov_num[col].std()).expandtabs(30))
```
