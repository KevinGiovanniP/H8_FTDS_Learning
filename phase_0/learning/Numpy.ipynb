**NOMOR 9**

Buatlah sebuah array seperti dibawah ini dan cari nilai `min`, `max`, dan `sum` untuk baris kedua dan kolom kedua.

```
m = [ [1, 2, 3], 
      [4, 5, 6],
      [7, 8, 9]]
```

````python
m = [ [1, 2, 3], 
      [4, 5, 6],
      [7, 8, 9]]

m = np.array(m)

m_baris = m.copy()
m_baris = m_baris[1:2]
print(m_baris.min(), m_baris.max(), m_baris.sum())

m_kolom = m.copy()
m_kolom = m_kolom[:, 1:2]
print(m_kolom.min(), m_kolom.max(), m_kolom.sum())
````
